Снимок экрана 2025-03-17 в 14.05.11.png

Решение:

```python
from math import sqrt

def calcul_formul(x):
  return (2**(2*x + 1) + 1) // 3

def isPrime(num):
  if num <= 1:
    return False
  for i in range(2, int(sqrt(num) + 1)):
    if num % i == 0:
      return False
  return True

for x in range(1, 37):
  num = calcul_formul(x)
  if isPrime(num):
    print(num)
```

Снимок экрана 2025-03-17 в 14.05.35.png

```python
from math import factorial

def calcForm(x):
  return factorial(x)/factorial(2**x)

i = 0
value_pred = calcForm(i)
sum = value_pred

while True:
  i+=1
  value_real = calcForm(i)
  if abs(value_pred - value_real) < 0.001:
    break
  sum+=value_real
  value_pred = value_real

print(sum)
```

Снимок экрана 2025-03-17 в 20.01.15.png

Решение:

```python
from math import gcd

def divide_frac(A, B, C, D):
  numerator = A * D
  denominator = B * C
  devisor = gcd(numerator, denominator)
  numerator = numerator / devisor
  denominator = denominator / devisor
  return numerator, denominator

if __name__=='__main__':
  A = int(input("Введи начальную точку A: "))
  B = int(input("Введи начальную точку B: "))
  C = int(input("Введи начальную точку C: "))
  D = int(input("Введи начальную точку D: "))
  print(divide_frac(A, B, C, D))
```
