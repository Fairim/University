```python
class Participant:
  def __init__(self, name, type_of_art):
    self.name = name
    self.typeArt = type_of_art

class Festival:
  def __init__(self):
    self.listParticipants = []

  def addParticipant(self, participant):
    self.listParticipants.append(participant)

  def showParticipant(self):
    for i in range(len(self.listParticipants)):
      print(f'Участник {i + 1}:')
      print(f'Имя: {self.listParticipants[i].name}')
      print(f'Вид исскуства: {self.listParticipants[i].typeArt}\n')


if __name__=='__main__':
  festival = Festival()
  partic = Participant('Денис', 'Программирования')
  festival.addParticipant(partic)
  partic = Participant('Данил', 'Писатель')
  festival.addParticipant(partic)
  festival.showParticipant()
```

```python
import abc
from math import pi

class D2Shape(abc.ABC):
  @abc.abstractmethod
  def square(self):
    pass
  @abc.abstractmethod
  def perimetr(self):
    pass

class Circle(D2Shape):
  def __init__(self, radius:float):
    self.radius = radius

  def square(self):
    return (self.radius ** 2) * pi

  def perimetr(self):
    return self.radius * 2 * pi

if __name__=='__main__':
  circle = Circle(3.5)
  print(f'Площадь круга: {circle.square()}')
  print(f'Площадь периметра: {circle.perimetr()}')
```

```python
class LeaveRequest:
  def __init__(self, employeeName, startDate, endDate):
    self.employeeName = employeeName
    self.startDate = startDate
    self.endDate = endDate
    self.is_approve = False

  def updateDates(self, localDateStart, localDateEnd):
    self.startDate = localDateStart
    self.endDate = localDateEnd

  def getRequestsInfo(self):
    return (f'Имя: {self.employeeName}\nДата: {self.startDate} - {self.endDate}')

  def approve(self):
    self.is_approve = True
    return (f"Заявка на отпуск сотрудника {self.employee_name} утверждена")


class PaidLeaveRequest(LeaveRequest):
  def __init__(self, employeeName, startDate, endDate):
    super().__init__(employeeName, startDate, endDate)

  def approve(self):
    self.is_approve = True
    return (f"Заявка на оплачиваемый отпуск сотрудника {self.employeeName} утверждена!")

class UnpaidLeaveRequest(LeaveRequest):
  def __init__(self, employeeName, startDate, endDate):
    super().__init__(employeeName, startDate, endDate)

  def approve(self):
    self.is_approve = True
    return (f"Заявка на не оплачиваемый отпуск сотрудника {self.employeeName} утверждена!")


if __name__=='__main__':
  paidLeave = PaidLeaveRequest('Данил', '13.03.2022', '13.04.2022')
  unpaidLeave = UnpaidLeaveRequest('Вадим', '15.01.2024', '25.03.2024')
  print(paidLeave.approve())
  print(unpaidLeave.approve())
```
