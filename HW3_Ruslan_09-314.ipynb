```python
def invert_dict(dict_A):
  dict_b = {}
  for key, value in dict_A.items():
    dict_b[value] = key
  return dict_b

dict_a = {'A': 2, 'b': 3, 'c':2, 'd':5, 'e':5}
print(invert_dict(dict_a))
```

Снимок экрана 2025-04-06 в 20.54.02.png

```python
def merge_dicts(dict_A, dict_B):
  merge_dict = dict_A
  for key, value in dict_B.items():
    if key in dict_A:
      merge_dict[key] += value
    else: merge_dict[key] = value
  return merge_dict

dict_A = {'A': 2, 'b': 3, 'c':2, 'd':5, 'e':5}
dict_B = {'A': 2, 'b': 3, 'y':7}
print(merge_dicts(dict_A, dict_B))
```

```python
def merge_values(dict_A, dict_B):
  merge_dict = {}
  for key in dict_A:
      merge_dict[key] = [dict_A[key]]
      if key in dict_B:
          merge_dict[key].append(dict_B[key])

  for key in dict_B:
      if key not in dict_A:
          merge_dict[key] = [dict_B[key]]
  return merge_dict

dict_A = {'A': 2, 'b': 3, 'c':2, 'd':5, 'e':5}
dict_B = {'A': 2, 'b': 3, 'y':7}
print(merge_values(dict_A, dict_B))
```

Снимок экрана 2025-04-06 в 22.40.29.png

```python
def sorted_tuple(tuple1):
  lst = list(tuple1)
  n = len(lst)
  for i in range(n):
    min_idx = i
    for j in range(i+1, n):
      if lst[j] < lst[min_idx]:
        min_idx = j
    lst[i], lst[min_idx] = lst[min_idx], lst[i]
  return lst

a = tuple([4,5,6,3,6,4,3,2.5,4,6])
print(sorted_tuple(a))
```
