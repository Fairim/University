```python
n = int(input("Введите размер массива: "))
mas = [i for i in range(n)]
memory = mas[n - 1]
for i in range(n-1, 0, -1):
  mas[i] = mas[i - 1]
mas[0] = memory
print(*(mas))
```

Снимок экрана 2025-04-01 в 11.28.54.png

```python
def return_list():
  dict_lead = {'A': 2, 'b': 3, 'c':2, 'd':5, 'e':5}
  list_read = []
  for key1, value1 in dict_lead.items():
    flag = True
    for key2, value2 in dict_lead.items():
      if value1 == value2 and key2 != key1:
        flag = False
        break
    if flag:
      list_read.append(value1)
  return list_read

print(*(return_list()))
```
