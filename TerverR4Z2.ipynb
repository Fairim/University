```python
import pandas as pd
import matplotlib.pyplot as plt

Y = 80

class LinRegres:
  def __init__(self, url):
    self.data = {'X': [], 'Y': []}
    self.read_file(url)
    self.len = len(self.data['X'])
    self.calc_med()
    self.S2x = self.med_x2 - (self.med_x ** 2)
    self.Dxy = (self.med_xy - self.med_x * self.med_y) / self.S2x
    self.c = self.med_x - self.Dxy * self.med_y
    print(self.calc_x(Y))
    self.draw_figure()

  def read_file(self, url):
    with open(url) as file:
      file.readline()
      for i in file:
        x, y = i.strip().split(',')
        self.data['X'].append(float(x))
        self.data['Y'].append(float(y))

  def calc_med(self):
    self.med_x = 0
    self.med_y = 0
    self.med_x2 = 0
    self.med_y2 = 0
    self.med_xy = 0
    for i in range(self.len):
      self.med_x += self.data['X'][i] / self.len
      self.med_y += self.data['Y'][i] / self.len
      self.med_x2 += (self.data['X'][i] ** 2) / self.len
      self.med_y2 += (self.data['Y'][i] ** 2) / self.len
      self.med_xy += (self.data['X'][i] * self.data['Y'][i]) / self.len

  def calc_x(self, y):
    return self.Dxy * y + self.c

  def draw_figure(self):
    min_y = min(self.data['Y'])
    max_y = max(self.data['Y'])
    plt.plot([self.calc_x(min_y), self.calc_x(max_y)], [min_y, max_y])
    for i in range(self.len):
      plt.scatter(self.data['X'][i], self.data['Y'][i], s=20, facecolors='none', edgecolors='black', linewidths=1)


LinRegres("r4z2.csv")
```
